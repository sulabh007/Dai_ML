```python
from sklearn.model_selection import KFold, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
from sklearn.calibration import LabelEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
import pandas as pd
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
```

```python
vehicle = pd.read_csv('Vehicle.csv')
lbl = LabelEncoder()
vehicle['Class'] = lbl.fit_transform(vehicle['Class'])
X = vehicle.drop('Class',axis=1)
y = vehicle['Class']
kfold =KFold(n_splits=5, shuffle=True, random_state=23)
lr = LogisticRegression()
knn = KNeighborsRegressor()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

```python
scaler = StandardScaler()
params = {'penalty':['l1','l2','elasticnet',None],
          'solver':['lbfgs', 'liblinear', 
                    'newton-cg', 'newton-cholesky', 'sag', 'saga'],
                    'multi_class':['ovr','multinomial']}


gcv = GridSearchCV(lr,param_grid=params,cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier


scaler = StandardScaler()
knn = KNeighborsClassifier()
pipe = Pipeline([('SCL',scaler),('KNN',knn)])

pipe.fit(X,y)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
```

```python


from sklearn.model_selection import GridSearchCV, StratifiedKFold


kfold=StratifiedKFold(n_splits=5, shuffle=True, random_state=23)

n_neighbors = np.arange(1,21)
print(pipe.get_params)
params = {'KNN__n_neighbors':n_neighbors,'SCL':[StandardScaler(),MinMaxScaler()]}
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```
