```python
import pandas as pd
import numpy as np
from sklearn.linear_model import  LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
```

```python
img = pd.read_csv('Vehicle.csv')
#Label Encoding
lbl = LabelEncoder()
img['Class'] = lbl.fit_transform(img['Class'])

X = img.drop('Class',axis=1)
y = img['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

Logistic Regression

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
lr = LogisticRegression()

penalty = ['l1','l2','elasticnet',None]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']

params = {'penalty':penalty,'solver':solver, 'multi_class':['ovr','multinomial']}
gcv = GridSearchCV(lr,param_grid=params,cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

KNN Classifier

```python
kfold=StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
knn = KNeighborsClassifier()
n_neighbors = np.arange(1,21)
scaler = StandardScaler()
pipe = Pipeline([('SCL',scaler),('KNN',knn)])
params = {'KNN__n_neighbors':n_neighbors,'SCL':[StandardScaler(),MinMaxScaler()]}
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

Gaussian NB

```python

from sklearn.model_selection import GridSearchCV, StratifiedKFold

nb = GaussianNB()
kfold=StratifiedKFold(n_splits=5, shuffle=True, random_state=23)

var_smoothing = np.linspace(0,1,20)

params = {'var_smoothing':var_smoothing}
gcv = GridSearchCV(nb,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
df = pd.DataFrame(gcv.cv_results_)
```

LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


lda = LinearDiscriminantAnalysis()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
params = {'solver':['svd', 'lsqr', 'eigen']}
gcv = GridSearchCV(lda, param_grid=params, cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

QDA

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis


qda = QuadraticDiscriminantAnalysis()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
params = {}
gcv = GridSearchCV(qda, param_grid=params, cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```
