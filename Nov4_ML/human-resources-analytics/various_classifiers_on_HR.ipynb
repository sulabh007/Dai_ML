```python
import pandas as pd
import numpy as np
from sklearn.linear_model import  LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
```

```python
HR = pd.read_csv('HR_comma_sep.csv')
dum_HR = pd.get_dummies(HR,drop_first=True)

X = dum_HR.drop('left',axis=1)
y = dum_HR['left']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2021)
```

Logistic Regression

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
lr = LogisticRegression()

penalty = ['l1','l2','elasticnet',None]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']

params = {'penalty':penalty,'solver':solver}
gcv = GridSearchCV(lr,param_grid=params,cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

KNN Classifier

```python
kfold=StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
knn = KNeighborsClassifier()
n_neighbors = np.arange(1,21)
scaler = StandardScaler()
pipe = Pipeline([('SCL',scaler),('KNN',knn)])
params = {'KNN__n_neighbors':n_neighbors,'SCL':[StandardScaler(),MinMaxScaler()]}
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

Gaussian NB

```python

from sklearn.model_selection import GridSearchCV, StratifiedKFold

nb = GaussianNB()
kfold=StratifiedKFold(n_splits=5, shuffle=True, random_state=23)

var_smoothing = np.linspace(0,1,20)

params = {'var_smoothing':var_smoothing}
gcv = GridSearchCV(nb,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
df = pd.DataFrame(gcv.cv_results_)
```
