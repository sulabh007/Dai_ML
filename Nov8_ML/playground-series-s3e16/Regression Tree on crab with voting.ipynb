```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.tree import  plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
from sklearn.ensemble import  VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.calibration import LabelEncoder


crab = pd.read_csv("train.csv",index_col=0)
lbl = LabelEncoder()
crab = pd.get_dummies(crab,drop_first=True)
X = crab.drop('Age',axis=1)
y = crab[['Age']]
kfold =KFold(n_splits=5, shuffle=True, random_state=23)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

```python
elastic = ElasticNet()
alphas = np.linspace(0.001, 5, 5)
l1_rat = np.linspace(0,1,5)
params = {'alpha':alphas,'l1_ratio':l1_rat}
gcv = GridSearchCV(elastic, param_grid=params, cv=kfold,verbose=3, scoring='r2')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
from sklearn.discriminant_analysis import StandardScaler


knn = KNeighborsRegressor()
scaler = StandardScaler()
pipe = Pipeline([('SCL',scaler),('KNN',knn)])
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
params = {'KNN__n_neighbors':[1,2,3,4],'SCL':[StandardScaler(),MinMaxScaler()]}
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold,verbose=3, scoring='r2', n_jobs=8)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
dtr = DecisionTreeRegressor(random_state=23, max_depth=None)
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
params = {'max_depth':[3,5,None],'min_samples_split':[2,5,10],'min_samples_leaf':[1,5,10]}
gcv = GridSearchCV(dtr, param_grid=params, cv=kfold,verbose=3, scoring='r2', n_jobs=8)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
voting = VotingRegressor(estimators=[('Elastic',elastic),('KNN',pipe),('DTR',dtr)])
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
params = {'DTR__max_depth':[3,5,None],'DTR__min_samples_split':[2,5,10],'DTR__min_samples_leaf':[1,5,10],
          'Elastic__alpha':alphas,'Elastic__l1_ratio':l1_rat,'KNN__KNN__n_neighbors':[1,2,3,4],'KNN__SCL':[StandardScaler(),MinMaxScaler()]}
gcv = GridSearchCV(voting, param_grid=params, cv=kfold, scoring='r2', n_jobs=8)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
voting = VotingRegressor(estimators=[('Elastic',elastic),('KNN',pipe),('DTR',dtr)])
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
params = {'DTR__max_depth':[3,5,None],'DTR__min_samples_split':[2,5,10],'DTR__min_samples_leaf':[1,5,10],
          'Elastic__alpha':alphas,'Elastic__l1_ratio':l1_rat,'KNN__KNN__n_neighbors':[1,2,3,4],'KNN__SCL':[StandardScaler(),MinMaxScaler()]}
gcv = GridSearchCV(voting, param_grid=params, cv=kfold,verbose=3, scoring='r2', n_jobs=8)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
from sklearn.model_selection import RandomizedSearchCV


voting = VotingRegressor(estimators=[('Elastic',elastic),('KNN',pipe),('DTR',dtr)])
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
params = {'DTR__max_depth':[3,5,None],'DTR__min_samples_split':[2,5,10],'DTR__min_samples_leaf':[1,5,10],
          'Elastic__alpha':alphas,'Elastic__l1_ratio':l1_rat,'KNN__KNN__n_neighbors':[1,2,3,4],'KNN__SCL':[StandardScaler(),MinMaxScaler()]}
gcv = RandomizedSearchCV(voting, param_distributions=params, cv=kfold, scoring='r2', n_jobs=8, n_iter=100, random_state=23)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
voting = VotingRegressor(estimators=[('Elastic',elastic),('KNN',pipe),('DTR',dtr)],weights=[0.55,0.48,0.53])
kfold = KFold(n_splits=5, shuffle=True, random_state=23)
params = {'DTR__max_depth':[3,5,None],'DTR__min_samples_split':[2,5,10],'DTR__min_samples_leaf':[1,5,10],
          'Elastic__alpha':alphas,'Elastic__l1_ratio':l1_rat,'KNN__KNN__n_neighbors':[1,2,3,4],'KNN__SCL':[StandardScaler(),MinMaxScaler()]}
rgcv = RandomizedSearchCV(voting, param_distributions=params, cv=kfold, scoring='r2', n_jobs=8, n_iter=100, random_state=23)
rgcv.fit(X,y)
print(rgcv.best_params_)
print(rgcv.best_score_)
```

```python
test = pd.read_csv('test.csv')
test = pd.get_dummies(test,drop_first=True)
bm = rgcv.best_estimator_

y_pred = bm.predict(test.iloc[:,1:])
print(y_pred)
submit = pd.DataFrame({'id':test.id,'Age':y_pred})
submit.to_csv('submission_for_Crab_using_voting.csv',index=False)
```
