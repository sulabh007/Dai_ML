```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
```

```python

img = pd.read_csv('Satellite.csv',sep=';')
#Label Encoding
lbl = LabelEncoder()
img["classes"] = lbl.fit_transform(img["classes"])

X = img.drop("classes",axis=1)
y = img["classes"]
```

```python
from sklearn.discriminant_analysis import StandardScaler
from sklearn.model_selection import KFold, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
```

```python


scaler = StandardScaler()
svm = SVC(kernel='rbf', probability=True, random_state=23)
kfold =KFold(n_splits=5, shuffle=True, random_state=23)
pipe = Pipeline([('SCL',scaler),('SVM',svm)])
kernel = ['linear','rbf']
params = {'SVM__C':np.linspace(0.001,5,100),'SVM__kernel':kernel, 'SVM__decision_function_shape':['ovo', 'ovr'],'SVM__gamma':np.linspace(0.001,5,100)}
rgcv = RandomizedSearchCV(pipe, param_distributions=params, cv=kfold, verbose=3, random_state=23, scoring='neg_log_loss')
rgcv.fit(X,y)
print(rgcv.best_params_)
print(rgcv.best_score_)
best_std_rbf = rgcv.best_estimator_
```

MinMaxScaler

```python
scaler = MinMaxScaler()
svm = SVC(kernel='linear', probability=True, random_state=23)
kfold =StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
pipe = Pipeline([('SCL',scaler),('SVM',svm)])
kernel = ['linear']
params = {'SVM__C':np.linspace(0.001,5,5),'SVM__kernel':kernel, 'SVM__decision_function_shape':['ovo', 'ovr'],'SVM__gamma':np.linspace(0.001,5,5)}
rgcv = RandomizedSearchCV(pipe, param_distributions=params, cv=kfold, verbose=3, random_state=23,  scoring='neg_log_loss')
rgcv.fit(X,y)
print(rgcv.best_params_)
print(rgcv.best_score_)
best_MM_rbf = rgcv.best_estimator_
```
