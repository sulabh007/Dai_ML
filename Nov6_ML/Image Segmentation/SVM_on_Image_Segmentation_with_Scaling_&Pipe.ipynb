```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
```

```python
img = pd.read_csv('Image_Segmention.csv')
#Label Encoding
lbl = LabelEncoder()
img['Class'] = lbl.fit_transform(img['Class'])

X = img.drop('Class',axis=1)
y = img['Class']
```

```python
from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC

stdscaler = StandardScaler()
svm = SVC(kernel='linear', probability=True, random_state=23)
pipe = Pipeline([('SCL',stdscaler),('SVM',svm)])
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
kernel = ['linear']
decision_function_shape = ['ovo', 'ovr']
params = {'SVM__C':np.linspace(0.001,5,10),'SVM__decision_function_shape':decision_function_shape}
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, verbose=3, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
minmaxscaler = MinMaxScaler()
svm = SVC(kernel='rbf', probability=True, random_state=23)
pipe = Pipeline([('SCL',minmaxscaler),('SVM',svm)])
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
kernel = ['rbf','linear']
gamma = np.linspace(0.0001,5,10)
decision_function_shape = ['ovo', 'ovr']
params = {'SVM__C':np.linspace(0.001,5,10),'SVM__decision_function_shape':decision_function_shape,'SCL':[StandardScaler(),MinMaxScaler()],
          'SVM__gamma':list(gamma)+['scale','auto'],'SVM__kernel':kernel}
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, verbose=3, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```
