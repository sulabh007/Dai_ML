```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
```

```python
img = pd.read_csv('Image_Segmention.csv')
#Label Encoding
lbl = LabelEncoder()
img['Class'] = lbl.fit_transform(img['Class'])

X = img.drop('Class',axis=1)
y = img['Class']
```

```python
from sklearn.svm import SVC


svm = SVC(kernel='linear', probability=True, random_state=23)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
kernel = ['linear']
decision_function_shape = ['ovo', 'ovr']
params = {'C':np.linspace(0.001,5,10),'decision_function_shape':decision_function_shape}
gcv = GridSearchCV(svm, param_grid=params, cv=kfold, verbose=3, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
from sklearn.svm import SVC


svm = SVC(kernel='rbf', probability=True, random_state=23)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=23)
kernel = ['rbf']
decision_function_shape = ['ovo', 'ovr']
params = {'C':np.linspace(0.001,5,10),'decision_function_shape':decision_function_shape}
gcv = GridSearchCV(svm, param_grid=params, cv=kfold, verbose=3, scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```
