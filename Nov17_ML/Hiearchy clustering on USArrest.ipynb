```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import StandardScaler
```

```python
USArrests = pd.read_csv('USArrests.csv', index_col=0)

scaler = StandardScaler()
USArrestsscaled = scaler.fit_transform(USArrests)
```

Calculating the linkages

```python
mergings = linkage(USArrestsscaled, method='average')
dendrogram(mergings, labels=list(USArrests.index),leaf_rotation=45, leaf_font_size=10)
plt.show()
```

sklearn

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

clustering = AgglomerativeClustering(n_clusters=2, linkage='average')
clustering.fit(USArrestsscaled)

print(clustering.labels_)
print(silhouette_score(USArrestsscaled, clustering.labels_))
```

```python
from sklearn.pipeline import Pipeline

clustering = AgglomerativeClustering(n_clusters=2, linkage='average')
pipe = Pipeline([('SCL', scaler), ('CLUST', clustering)])
pipe.fit(USArrests)
USArrestsscaled = scaler.fit_transform(USArrests)
print(clustering.labels_)
print(silhouette_score(USArrestsscaled, clustering.labels_))
```

```python
clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
pipe = Pipeline([('SCL', scaler), ('CLUST', clustering)])
pipe.fit(USArrests)
USArrestsscaled = scaler.fit_transform(USArrests)
print(clustering.labels_)
print(silhouette_score(USArrestsscaled, clustering.labels_))
```

```python
clustering = AgglomerativeClustering(n_clusters=2, linkage='single')
pipe = Pipeline([('SCL', scaler), ('CLUST', clustering)])
pipe.fit(USArrests)
USArrestsscaled = scaler.fit_transform(USArrests)
print(clustering.labels_)
print(silhouette_score(USArrestsscaled, clustering.labels_))
```

```python
clustering = AgglomerativeClustering(n_clusters=2, linkage='complete')
pipe = Pipeline([('SCL', scaler), ('CLUST', clustering)])
pipe.fit(USArrests)
USArrestsscaled = scaler.fit_transform(USArrests)
print(clustering.labels_)
print(silhouette_score(USArrestsscaled, clustering.labels_))
```
