# PCA

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
milk = pd.read_csv("Datasets/milk.csv",index_col=0)
```

```python
scalar = StandardScaler()
scaled_data = scalar.fit_transform(milk)
prcomp = PCA()

score = prcomp.fit_transform(scaled_data)

v_covar = np.cov(scaled_data)
values, vector = np.linalg.eig(v_covar)

print("Eigen Values : ",values)
print("Variance Of PC Scores calculated manually : ",np.var(score, axis=0,ddof=1))
print(prcomp.explained_variance_)
```

percentage of variation explained

```python
tot_var = np.sum(prcomp.explained_variance_)
print("Percentage Of Age Variation : ")
print((prcomp.explained_variance_/tot_var)*100)
```

adding first 3 percentage of variation explained suggest that first 3 columns contributes about 97.8 percent information

percentage of variation explained

```python
print("Cumulative Percentage Age Variation Explaned : ")
print(np.cumsum(prcomp.explained_variance_ratio_*100))
```
